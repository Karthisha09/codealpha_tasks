```python
train = train.drop(['Ticket'], axis=1) 
test = test.drop(['Ticket'], axis=1)
```

```python
train = train.drop(['Cabin'], axis=1) 
test = test.drop(['Cabin'], axis=1)
```

```python
train = train.fillna({"Embarked": "S"}) 
```

```python
mean_age = data['Age'].mean()
data['Age'] = data['Age'].fillna(mean_age)
```

```python
sex_mapping = {"male": 0, "female": 1} 
train['Sex'] = train['Sex'].map(sex_mapping) 
test['Sex'] = test['Sex'].map(sex_mapping) 
  
embarked_mapping = {"S": 1, "C": 2, "Q": 3} 
train['Embarked'] = train['Embarked'].map(embarked_mapping) 
test['Embarked'] = test['Embarked'].map(embarked_mapping) 
```

```python
from sklearn.model_selection import train_test_split 
predictors = train.drop(['Survived', 'PassengerId'], axis=1) 
target = train["Survived"] 
x_train, x_val, y_train, y_val = train_test_split(predictors, target, test_size=0.3, random_state=0) 
```

```python
x_train.info()
```

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy='mean')
x_train_imputed = imputer.fit_transform(x_train)
```

```python
from sklearn.ensemble import RandomForestClassifier 
from sklearn.metrics import accuracy_score 
  
randomforest = RandomForestClassifier(random_state=0) 
randomforest.fit(x_train_imputed, y_train) 
```

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy='mean')
x_val_imputed = imputer.fit_transform(x_val)
y_val_pred = randomforest.predict(x_val_imputed)

acc_randomforest = round(accuracy_score(y_val_pred, y_val) * 100, 2)
print(f"Accuracy on validation set: {acc_randomforest}%")
```
